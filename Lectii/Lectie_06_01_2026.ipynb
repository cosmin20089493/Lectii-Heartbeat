```python
import re

text = "Suna-ma la 0722-123-456 sau la 0744-999-000."
# Sablon: 4 cifre, cratima, 3 cifre, cratima, 3 cifre
sablon = r"\d{4}-\d{3}-\d{3}"

rezultate = re.findall(sablon, text)
print(rezultate)
# Output: ['0722-123-456', '0744-999-000']
```
