```python
#Buble sort

lst = [ 2, 5, 7, 9 ,21 , 1, 10]
minim = lst[0]   #variabila auxiliara

for x in lst[1:]:     #-> actiune
    if x < minim:      #-> actiune
        minim = x
        print("Am gasit valoarea minima dar nu o pot afisa")
    minim = x
    print(f"Afisez un minim  ales de mine aleatoriu", minim )
    print ("Am terminat lista")


print(f"Cea mai mica valoarea este :" ,minim )
```

```python
#bubble sort

n = len(lst)
for i in range(n - 1):
    swapped = False
    for j in range(n - 1 - i):
        if lst[j] > lst[j + 1]:
            lst[j], lst[j + 1] = lst[j + 1], lst[j]
            swapped = True
    if not swapped:
        break

print("Lista sortata:", lst)
print("Minimul (pozitia 0):", lst[0])
```
