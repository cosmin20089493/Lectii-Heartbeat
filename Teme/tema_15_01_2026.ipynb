## Partea A — Listă simplu înlănțuită (SLL)


  N = cate numere are lista

urmatorul input = exact N numere

X = valoarea care trebuie stearsa (prima aparitie)



```python
# Tema: lista simplu inlantuita (SLL) fara class/def/import
# Simulez nodurile cu 2 liste: val si nxt


N = int(input())
print("Am citit N =", N)



values = []
while len(values) < N:
    parts = input().split()
    for p in parts:
        if len(values) < N:
            values.append(int(p))

X = int(input())

val = []
nxt = []
head = -1
tail = -1

# A1) construiesc lista prin inserare la final
i = 0
while i < N:
    v = values[i]

    val.append(v)
    nxt.append(-1)
    new_idx = len(val) - 1

    if head == -1:
        head = new_idx
        tail = new_idx
    else:
        nxt[tail] = new_idx
        tail = new_idx

    i += 1
print("Valori citite:", values)

# A2) afisez lista
cur = head
text = ""
while cur != -1:
    if text == "":
        text = str(val[cur])
    else:
        text = text + " -> " + str(val[cur])
    cur = nxt[cur]

print("Lista initiala: " + text)

# A3) suma
suma = 0
cur = head
while cur != -1:
    suma = suma + val[cur]
    cur = nxt[cur]

print("Suma: " + str(suma))

# A4) sterg primul nod cu valoarea X
prev = -1
cur = head

while cur != -1 and val[cur] != X:
    prev = cur
    cur = nxt[cur]

if cur == -1:
    # nu am gasit X, nu fac nimic
    pass
else:
    # am gasit nodul cur
    if prev == -1:
        # sterg head
        head = nxt[head]
        if head == -1:
            tail = -1
    else:
        # sar peste cur
        nxt[prev] = nxt[cur]
        if cur == tail:
            tail = prev

# afisez lista dupa stergere
cur = head
text2 = ""
while cur != -1:
    if text2 == "":
        text2 = str(val[cur])
    else:
        text2 = text2 + " -> " + str(val[cur])
    cur = nxt[cur]

print("Lista dupa stergere: " + text2)
```
