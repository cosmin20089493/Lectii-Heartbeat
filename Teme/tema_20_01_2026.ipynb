# 1. TIPURI DE ALGORITMI – CLASIFICARI
# =========================================================


# a) Algoritmi simpli vs algoritmi eficienti
#
# Algoritmi simpli:
# - usor de inteles
# - usor de implementat
# - lenti pentru liste mari (complexitate mare)
# Exemple: Bubble Sort, Selection Sort, Insertion Sort
#
# Algoritmi eficienti:
# - mai rapizi pentru date multe
# - folosesc idei mai avansate
# - complexitate mai buna
# Exemple: Merge Sort, Quick Sort


# b) In-place vs extra-memorie
#
# In-place:
# - sorteaza direct lista
# - nu folosesc liste auxiliare mari
# Exemple: Bubble, Selection, Insertion, Quick Sort
#
# Extra-memorie:
# - folosesc liste ajutatoare
# - consuma mai multa memorie
# Exemplu: Merge Sort


# c) Stabili vs instabili
#
# Stabil:
# - elementele egale isi pastreaza ordinea initiala
#
# Instabil:
# - elementele egale pot schimba ordinea

# 6. ALGORITMI DE SORTARE – IMPLEMENTARI SIMPLE
# =========================================================


```python
# -------------------------
# Bubble Sort
# -------------------------

def bubble_sort(a):
    a = a[:]
    n = len(a)
    for i in range(n):
        for j in range(0, n - 1 - i):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a

print(bubble_sort([3, 1, 2]))
```

```python
# -------------------------
# Selection Sort
# -------------------------

def selection_sort(a):
    a = a[:]
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a

print(selection_sort([3, 1, 2]))
```

```python
# -------------------------
# Insertion Sort
# -------------------------

def insertion_sort(a):
    a = a[:]
    for i in range(1, len(a)):
        x = a[i]
        j = i - 1
        while j >= 0 and a[j] > x:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = x
    return a

print(insertion_sort([3, 1, 2]))
```

```python
# -------------------------
# Merge Sort
# -------------------------

def merge(left, right):
    i = j = 0
    out = []
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            out.append(left[i])
            i += 1
        else:
            out.append(right[j])
            j += 1
    out.extend(left[i:])
    out.extend(right[j:])
    return out

def merge_sort(a):
    if len(a) <= 1:
        return a
    mid = len(a) // 2
    return merge(merge_sort(a[:mid]), merge_sort(a[mid:]))

print(merge_sort([3, 1, 4, 2]))
```

```python
# -------------------------
# Quick Sort (simplificat, nu in-place)
# -------------------------

def quick_sort(a):
    if len(a) <= 1:
        return a
    pivot = a[0]
    left = [x for x in a[1:] if x <= pivot]
    right = [x for x in a[1:] if x > pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)

print(quick_sort([3, 1, 4, 2]))
```

# 7. DIFERENTA DINTRE MERGE SORT SI QUICK SORT
# =========================================================


# Merge Sort:
# - divide et impera
# - O(n log n) garantat
# - stabil
# - foloseste memorie extra
#
# Quick Sort:
# - divide et impera
# - O(n log n) in medie, O(n^2) in cel mai rau caz
# - instabil
# - in-place (varianta clasica)
#
# In practica:
# - Merge Sort: sigur, dar mai multă memorie
# - Quick Sort: foarte rapid, dar depinde de pivot
