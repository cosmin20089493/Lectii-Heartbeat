# Ce reprezinta un virual environment (venv) ?


Conceptual: ce reprezinta
    Avem un folder al proiectului si pe langa cod mai avem o "cutie" cu :

-pip-ul acestui proiect

-biblioteciile instalate pentru acest proiect

-referinte catre interpretorul Python

  Cand  "activam " mediul, comanda python si pip vor folosii cutia aceea, nu sistemul.

D**e ce se foloseste daca putem rula direct in Python ?**

Daca instalezi totul "global":

-doua proiecte pot cere versiuni diferite din aceleasi librarie (ex: requests, numpy...) si apar conflictele.

-iti "murdaresti" sistemult si nu mai stii ce ai instalat si pentru ce

-nu mai poti reproduce usor proiectul pe alt PC/server (sau peste ceva timp)

**Cu venv:**

-fiecare proiect are dependentele lui

-update-urile nu strica alte proiecte

-avem instalari curate si reproductibile ( de obicei in fisierul requirements.txt)

### Diferente intre venv si Conda

**VENV** :

   -izoleaza pachetele Python (instalate cu pip)

   -foloseste de regula un Python deja instalat ( poate crea mediu cu acel interpretor)

   -e standard si usir de folosit in proiecte obijnuite Python

**CONDA**:

   -e si "manager de medii" si "manager de pachete"

   -poate instala si alte dependente non Python (biblioteci c/C++, MKL...)

   -poate crea medii cu versiuni diferite de Python foarte usor

   -e popular in data science, ML, cand ai dependente native mai sesnibile

## Comentariu pentru Bubble sort

```python
n = len(lst)      #nr de elemente din lista
for i in range(n - 1):  # i este un contor de ture (0,1,2,...n-1)

    swapped = False  #presupunem ca tura asta nu facem nici un schimb, daca ramane False , lista e deja soratata.
    for j in range(n - 1 - i):  #j este indexul perechii comparate. n-1-i inseamna ca la fiecare tura i , nu mai verificam ultimul element fiindca la finalul fiecarei ture cel mai mare "urca" la capat si e deja la locul lui
        if lst[j] > lst[j + 1]:  #comapram doua elemente vecine , cel de pe pozita j cu cel de pe pozitia j+1
            lst[j], lst[j + 1] = lst[j + 1], lst[j]  #aici se face swap, schimb de locuri
            swapped = True   # marcam ca am facut macar un schimb in tura asta
    if not swapped:   #daca nu exista nici un swap intr-o tura completa, lista e deja sortata, deci ne oprim.
        break

print("Lista sortata:", lst)
print("Minimul (pozitia 0):", lst[0])
```
