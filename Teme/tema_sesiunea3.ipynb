## 1.Creează o buclă care cere un PIN de 4 cifre până e corect

```python
while True:
    pin = input("Seteaza un pin de 4 cifre: ")    #string pentru a putea utilza si 0 in fata
    if len(pin) == 4 and pin.isdigit() :         # len verifica lungimea iar isdigit verifica sa fie cifre
        break                      # folosim break pentru a nu face o bucla infinita daca conditia eset adevarata
    print("PIN invalid. Trebuie fix 4 cifre (ex:0123)." )

pin_corect = input("Introdu codul pin de 4 cifre: ")
if pin_corect !=pin:
    print("Accesul respins")
else:
    print("Accesul permis")
```

## 3.Creează o piramidă inversă cu for


```python
'''
LOgica din spate , unde i =5
Linia 1 — i = 5

Spații: range(0, 5-5) = range(0, 0) → 0 spații

Stele: range(0, 5) → j = 0,1,2,3,4 → 5 ori „* ”

Ecran:

* * * * *

Linia 2 — i = 4

Spații: range(0, 5-4) = range(0, 1) → 1 spațiu

Stele: range(0, 4) → 4 ori „* ”

Ecran:

 ␣* * * *

Linia 3 — i = 3

Spații: range(0, 5-3) = range(0, 2) → 2 spații

Stele: range(0, 3) → 3 ori „* ”

Ecran:

 ␣␣* * *

Linia 4 — i = 2

Spații: range(0, 5-2) = range(0, 3) → 3 spații

Stele: range(0, 2) → 2 ori „* ”

Ecran:

 ␣␣␣* *

Linia 5 — i = 1

Spații: range(0, 5-1) = range(0, 4) → 4 spații

Stele: range(0, 1) → 1 dată „* ”

Ecran:

 ␣␣␣␣*


'''
marime = int(input("Introdu cate linii sa aibe piramida inversa : "))
for i in range(marime , 0, -1):
    for j in range(0,marime-i):
        print(end=" ")
    for j in range (0,i):
        print("*", end=" ")
    print()
```

## 4.Scrie o buclă care adună toate numerele pare între 1 și 100

```python
'''
Am luat in calcul si numarul 100, enuntul poate fi interpretat in doua feluri, cu sau fara 100. Daca vrem fara 100 schimbam in range, in loc de 101, scriem 100.
'''
suma = 0                    #variabila pentru stocarea sumei
for i in range(2, 101, 2):
    suma +=i                 # Adaugam fiecare numar la suma totala ( mai exact: 0+2+4+...+100)
print(f"Suma numerelor pare este: {suma}")
```
