## 2.Creează un tuplu cu 4 coordonate GPS și afișează fiecare pe linie.

```python
coordonate = (
    (45.7580, 21.2290),  # Timisoara
    (44.4268, 26.1025),  # Bucuresti
    (47.4979, 19.0402),  # Budapesta
    (51.1657, 10.4515)   # Germania
)

for lat, lon in coordonate:
    print(f"Latitudine: {lat}, Longitudine: {lon}")



"""1.for ia pe rând fiecare element din tuplul mare. Fiecare element este un tuplu de forma (45.7580, 21.2290) etc.

2.„lat, lon” este despachetare (tuple unpacking). Python încearcă să asocieze primul element al tuplului curent cu lat și al doilea cu lon.
     • la prima iterație: lat = 45.7580, lon = 21.2290
    • la a doua: lat = 44.4268, lon = 26.1025  … și așa mai departe.

3.Apoi execuți print pentru perechea curentă, treci la următoarea iterație, iar variabilele lat, lon primesc noile valori.

Detalii utile:

        Ordinea contează: primul element merge în lat, al doilea în lon.

        Numărul variabilelor din stânga trebuie să egaleze numărul elementelor din tuplul iterat; altfel apare ValueError.

        Iterezi o structură imuabilă (tuplu), dar despachetarea creează variabile noi la fiecare pas.
        """;

```

## 4.Creează dicționar produse:preț → afișează doar cele > 50 lei
