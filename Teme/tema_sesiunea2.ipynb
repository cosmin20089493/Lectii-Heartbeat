```python
Exercițiu:

Simulează un scor la test cu calificative:

10: "Excelent"

8-9: "Foarte bine"

6-7: "Bine"

5: "Trecut"

<5: "Respins"
```

```python

scor_test= int(input("Ce scor ai obtinut?:"))

if scor_test < 5:
    print("Respins!")
elif scor_test == 5:
    print("Trecut.")
elif  scor_test == 6:
    print("Bine.")
elif scor_test == 7:
    print("Bine.")
elif scor_test ==8:
    print("Foarte bine.")
elif scor_test == 9:
    print("Foarte bine.")
else:
    print("Excelent.")
```
