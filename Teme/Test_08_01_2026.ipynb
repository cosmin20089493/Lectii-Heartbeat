6.Break opreste fortat executia si trece la urmatoarea instructiune.Continue sare peste restulu codului din iteratia curenta si trece direct la urmatoarea repetare.

```python
#P1) Suma cifrelor

n = int(input("Introdu un numar: "))
suma = 0

while n > 0:
    suma += n % 10  # adaugă ultima cifră
    n //= 10         # elimină ultima cifră

print(suma)
```

```python
#P2) Inversarea unui numar (12p)

n = int(input("Introdu un numar: "))
invers = 0

while n > 0:
    invers = invers * 10 + n % 10  # n%10 = extrage ultima cifra, iar
    n //= 10  #elimina ultima cifra din numarul initial

print(invers)
```

```python
# 3) Divizorii unui numar (12p)


n = int(input("Introdu un numar: "))

for i in range(1, n + 1):  #generam toate nr de la 1 pana la n
    if n % i == 0:        #verificam daca i este divizor afland daca restul impartirii lui n la i este 0.
      print(i,  end=",")    #end" " punem virgula intre numere in loc sa trecem la rand nou


```

```python
#P4) Secunde -> ore:minute:secunde (12p)

s = int(input("Introdu un numar aleator de secunde: "))

h = s // 3600
m = (s % 3600) // 60
secunde_rest = s % 60

print(f"{h}:{m}:{secunde_rest}")
```
