### 🔹 Ce este o funcție (def)

O funcție este o bucată de cod care are un nume și face o treabă clară.
O scrii o singură dată, apoi o poți folosi de oricâte ori vrei.

### 🔹 Cum arată o funcție

def salut():
    print("Salut, Cosmin!")


def — vine de la define, adică „definește o funcție”.

salut — este numele funcției.

### () — pot conține datele de intrare (argumente).

: — marchează începutul codului din funcție.

Codul din interior se scrie indentat (cu 4 spații).

Pentru a o folosi:

salut()  # apelează funcția

### 🔹 Funcție cu parametri (date de intrare)

Dacă vrei ca funcția să lucreze cu ceva ce-i dai tu:

def salut(nume):
    print(f"Salut, {nume}!")


→ o apelezi așa:

salut("Maria")
salut("Ion")

### 🔹 Funcție care întoarce un rezultat (return)

def aduna(a, b):
    return a + b


return trimite rezultatul înapoi acolo unde a fost apelată funcția.

Poți salva acel rezultat:

suma = aduna(5, 3)
print(suma)   # 8

### 🔹 De ce sunt utile

Poți reutiliza codul fără să-l scrii de fiecare dată.

Codul devine curat și organizat.

Poți face funcții pentru orice: calcule, mesaje, afișări, verificări etc.

```python
angajati = []  #lista goala care va tine toti angajatii

def adauga_angajat(nume, salariu, pozitie):   #definim o functie cu 3 parametrii:nume(string), salariu(numar), pozitie(strig).
    angajat = {"nume": nume, "salariu": salariu, "pozitie": pozitie} #construim un dictionare pentru un angajat cu 3 chei: nume, salariu, pozitie.
    angajati.append(angajat) #adaugam dictionarul in lista globala angajati

def afiseaza_tot():  #functie care afiseaza toti angajatii din lista
    for a in angajati:   # parcurgem lista; a este fiecare dictionar-angajat pe rand
        print(f"{a['nume']} - salariu: {a['salariu']} - pozitie: {a['pozitie']}") # Afisam valorile din dictionarul a. f"..." este un f-string: insereaza valoriile direct in text.

def calculeaza_media(): # functie care calculeaza salariul mediu din toti angajatii
    if not angajati:   #este o protectie in caz ca lista este goala , intoarce 0 ca sa evitam impartirea la 0
        return 0
    total = sum(a['salariu'] for a in angajati) # ia pe rand ["salariu"] pentru fiecare angajat .... sum(...) aduna toate salariile intr-un total
    return total / len(angajati)   # media= totalul impartit la numarul de angajati , aici ne folosim de len ca sa numere angajatii.

def afiseaza_salariul_mediu():  #functie care calculeaza si afiseaza media
    media = calculeaza_media()
    print(f"Salariul mediu este: {media:.2f} lei")  #afisam media cu doua zecimale ( :2f formateaza numarul cu doua zecimale)

def salariu_peste_medie():   #functie care afiseaza angajatii cu salariu > media
    media = calculeaza_media() #calculam o singura data media de referinta
    print(f"Salariul mediu este: {media:.2f} lei")
    print("Angajații cu salariu peste medie:")
    for a in angajati:   #parcurgem fiecare angajat a
        if a['salariu'] > media:  #doar cine are salariul >media trece mai departe
            print(f"{a['nume']} - {a['salariu']} lei ({a['pozitie']})") #afisam nume, salariu si pozitia pentrz acei angajati.

# Test:
adauga_angajat("Andrei Pop" , 6200, "Programator")
adauga_angajat("Maria Ionescu" , 8500, "Manager")
adauga_angajat("Ion Georgescu" , 4200, "Tehnician")
adauga_angajat("Elena Radu" , 4800, "Contabil")
adauga_angajat("Cristian Dima" , 5500, "Mecanic")



salariu_peste_medie()
```


## 2. Text Analytics

### Funcție care primește text → returnează:

### nr. de cuvinte

### cel mai lung cuvânt

### dacă apare „python”

```python
text = input("Scrie un text: ")   # scriem un text de tip string

def nr_cuvinte(text) :  #functie unde memoram numarul de cuvintele
    cuvinte = text.split()    # functia split imparte textul in lista de cuvinte
    return len(cuvinte)  #intoarce nr de cuvinte
nr =nr_cuvinte(text)   #apelul functiet

def lungime_cuvant(text):
    cuvinte = text.split()
    cel_mai_lung_cuvant  = ""
    for cuvant in cuvinte:
        if len(cuvant) > len(cel_mai_lung_cuvant):
            cel_mai_lung_cuvant = cuvant
    return cel_mai_lung_cuvant


rezultat = lungime_cuvant(text)

def contine_python(text):
    tex_mic =text.lower()   #transformam tot textul in litere mici.
    if "python" in tex_mic:  #verificam daca apare cuvntul python cu if.
        return True
    else:
        return False

contine = contine_python(text)     #apelul functiei



print(f"In propozitie sunt {nr} cuvinte")
print(f"Cel mai lung cuvant este : {rezultat}")
print(f"Apare cuvantu PYTHON ?: {contine}")


```


## 3. Coș de cumpărături

### Funcție adaugă produse

### Funcție total cu TVA

### Funcție bon fiscal cu print()

```python
cos = []   # lista de produse (fiecare produs = dict {"nume": str, "pret": float})

def adauga_produs(nume: str, pret: float):
    cos.append({"nume": nume, "pret": pret})

def total_cu_tva(tva: float = 0.19) -> float:
    total_fara = sum(p["pret"] for p in cos)
    return total_fara * (1 + tva)

def bon_fiscal(tva: float = 0.19):
    if not cos:
        print("Cosul este gol.")
        return
    print("\n=== BON FISCAL ===")
    for p in cos:
        print(f"{p['nume']:<25} {p['pret']:>8.2f} lei")
    total_fara = sum(p["pret"] for p in cos)
    valoare_tva = total_fara * tva
    total_cu = total_fara + valoare_tva
    print("-" * 36)
    print(f"{'Subtotal:':<25} {total_fara:>8.2f} lei")
    print(f"{'TVA (19%):':<25} {valoare_tva:>8.2f} lei")
    print(f"{'TOTAL:':<25} {total_cu:>8.2f} lei")
    print("===================\n")

def meniu():
    while True:
        print("1. Adauga produs")
        print("2. Afiseaza totalul cu TVA")
        print("3. Tipareste bon fiscal")
        print("0. Iesire")
        opt = input("Alege: ").strip()

        if opt == "1":
            nume = input("Nume produs: ").strip()
            pret_txt = input("Pret (lei): ").strip().replace(",", ".")
            try:
                pret = float(pret_txt)
                if pret < 0:
                    print("Pretul nu poate fi negativ.")
                    continue
                adauga_produs(nume, pret)
                print("Produs adaugat.")
            except ValueError:
                print("Pret invalid. Incearca din nou.")
        elif opt == "2":
            total = total_cu_tva()
            print(f"Total cu TVA: {total:.2f} lei")
        elif opt == "3":
            bon_fiscal()
        elif opt == "0":
            print("La revedere!")
            break
        else:
            print("Optiune invalida.")

if __name__ == "__main__":
    meniu()
```
