## 2. Coin Change – testare

Apeleaza functia cu:

sume diferite

seturi diferite de monede

Observa unde da rezultate bune si unde nu

```python
def coin_change(coins, amount):
    coins.sort(reverse=True)
    result = []
    for coin in coins:
        while amount >= coin:
            amount -= coin
            result.append(coin)
    return result

print(coin_change([1, 5, 10, 25], 83))  # 83 = 25+25+25+5+1+1+1

print(coin_change([1, 2, 5], 10))       # 10=5+5

print(coin_change([1, 5, 10], 18))      # 18=10+5+1+1+1

print(coin_change([1, 3, 4], 6))   # Aici rezultatul este [4, 1, 1] desii solutia optima este  [3 ,3]. Se intampla asta deoarece algoritmul greedy NU garanteaza solutia optima pentru orice set de monede, mai exact, ia o decizie locala (cea mai mare moneda acum) si nu se uita in "viitor".





```

```python

'Aici sincer nici nu am inteles cerinta exacta, habar nu am ce se cere, nici chat gpt nu ma facut sa inteleg exact...dar in schimb mi a rezolvat problema. cu toate ca nu stiu ce afiseaza si de ce si cum functioneaza'

def sigiliu_corect(s: str) -> bool:
    perechi = {')': '(', ']': '[', '}': '{', '>': '<'}
    stiva = []

    for ch in s:
        if ch in "([{<":
            stiva.append(ch)
        else:
            # ch este una din parantezele de inchidere
            if not stiva:
                return False
            top = stiva.pop()
            if perechi[ch] != top:
                return False

    return len(stiva) == 0


def test_sigiliu_corect():
    tests = [
        ("([])<{}>", True),
        ("([)]", False),
        ("<<>>({[]})", True),
        ("(((", False),
        ("", True),
    ]
    for s, expected in tests:
        got = sigiliu_corect(s)
        print("sigiliu_corect(", repr(s), ") =", got, " expected=", expected)

print(sigiliu_corect("([])<{}>"), "expected True")
print(sigiliu_corect("([)]"), "expected False")
print(sigiliu_corect("<<>>({[]})"), "expected True")
print(sigiliu_corect("((("), "expected False")
print(sigiliu_corect(""), "expected True")
```

## Papirusul “Top 3 Bârfe” (HashMap)

```python
def top_3_barfe(text: str):
    # 1) lower-case
    t = text.lower()

    # 2) inlocuire punctuatie
    for p in ".,!?;:":
        t = t.replace(p, " ")

    # 3) frecventa cuvintelor
    frec = {}
    for w in t.split():
        frec[w] = frec.get(w, 0) + 1

    rezultat = []
    folosite = {}

    # alegem exact top 3 (sau mai putine daca nu sunt)
    for _ in range(3):
        best_word = None
        best_count = -1

        for w in frec:
            if w in folosite:
                continue
            c = frec[w]

            if (best_word is None or
                c > best_count or
                (c == best_count and w < best_word)):
                best_word = w
                best_count = c

        if best_word is None:
            break

        rezultat.append(best_word)
        folosite[best_word] = True

    return rezultat


#textint = int(input("Introdu barfele"))
#text=textint;

text = "Nisip, nisip! Soare... nisip? Faraon, faraon!, faraon, soare ,pisica , catel , caltel, catel, pisica"
print(top_3_barfe(text))

```

## Cămila cu Inventar (Sort + Binary Search)

```python

#1.Selection sor




def selection_sort(a):
    # lungimea listei
    n = len(a)

    # mergem pe lista de la stanga la dreapta
    for i in range(n):

        # presupunem ca elementul de pe pozitia i este minim
        min_index = i

        # cautam in partea din dreapta un element mai mic
        for j in range(i + 1, n):
            if a[j] < a[min_index]:
                min_index = j

        # daca am gasit un minim mai mic, il schimbam
        a[i], a[min_index] = a[min_index], a[i]

    # returnam lista sortata
    return a

a = [7, 2, 9, 2, 5, 2]

selection_sort(a)  #aici facem sortarea

print("Lista sortata:", a)

```

```python
#2.Binary search

def binary_search(a, x):
    # capatul din stanga
    stanga = 0
    # capatul din dreapta
    dreapta = len(a) - 1

    # cat timp intervalul este valid
    while stanga <= dreapta:

        # calculam mijlocul
        mijloc = (stanga + dreapta) // 2

        # daca am gasit valoarea cautata
        if a[mijloc] == x:
            return mijloc

        # daca valoarea din mijloc e mai mica
        # cautam in dreapta
        if a[mijloc] < x:
            stanga = mijloc + 1
        else:
            # altfel cautam in stanga
            dreapta = mijloc - 1

    # daca nu am gasit nimic
    return -1

a = [7, 2, 9, 2, 5, 2]

lista_sortata = a[:]        # copie
selection_sort(lista_sortata)

print("Lista sortata:", lista_sortata)
print("Caut 5:", binary_search(lista_sortata, 5))
print("Caut 8:", binary_search(lista_sortata, 8))
```
