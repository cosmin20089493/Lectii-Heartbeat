
## 1. Suma primelor `n` numere naturale

Exemplu:

n = 5 → Output: 15 (1 + 2 + 3 + 4 + 5)

```python
n = int(input("Introdu un numar natural (pozitiv): "))

if n < 0:
    print("Numarul nu este pozitiv, incearca din nou ")

else:
    rezultat = n * ( n +1 ) // 2
    print(f"Suma primelor {n} numere naturale este: {rezultat}")
```

4. Inversarea unui număr întreg

Exemplu:

n = 1234 → Output: 4321

```python
n = int(input("Introdu un numar natural (pozitiv): "))
invers = 0

while n > 0:
    ultima_cifra = n % 10          # Extrage ultima cifra (4)
    invers = invers * 10 + ultima_cifra  # O adaugă la noul numar
    n = n // 10                    # Taie ultima cifra din n original

print(invers)
# Output: 4321
```


## 5. Verifică dacă un cuvânt este palindrom

Exemplu:

text = "ana" → Output: True

text = "python" → Output: False

```python
text=input("Introdu un text: ")

def verifica_palindrom(text):
    # Transformam in litere mici și eliminam spatiile
    text_curat = text.lower().replace(" ", "")
    return text_curat == text_curat[::-1]

print(verifica_palindrom(text))
```


## 12. Află suma cifrelor unui număr

Exemplu:

n = 123 → Output: 6



```python
n = int(input("Introdu un numar natural: "))
suma = 0

while n > 0:
    cifra = n % 10    # Luam ultima cifra
    suma += cifra     # O adunăm la suma
    n = n // 10       # Taiem ultima cifra

print(f"Suma cifrelor este: {suma}")
# Output: 6
```


## 14. Află aria unui cerc de rază `r`

Exemplu:

r = 3 → Output: 28.27 (cu 2 zecimale)



```python
#Formula matematica pentru aria unui cerc : A = π · r²

r = int(input("Introdu raza r= : "))
pi_aproximativ = 3.14159

aria = pi_aproximativ * r * r
print(f"Aria cercului este:{aria:.2f}")  #{aria:.2f}: Aceasta este o metodă de formatare a textului (f-string). Partea .2f îi spune lui Python: "afișeaza acest numar de tip float (zecimal) cu exact 2 cifre dupa punct".
```

```python
n = int(input("Introdu un numar natural: "))
n_original = n # pastram o copie pentru afisare

suma_pare = 0
suma_impare = 0

while n > 0:
    cifra = n % 10
    if cifra % 2 ==0:
        suma_pare += cifra
    else:
        suma_impare += cifra
    n = n // 10

este_norocos = (suma_pare ==suma_impare )

print(f"Suma pare: {suma_pare}, Suma impare: {suma_impare}")
print(f"Este numarul {n_original} norocos? : {este_norocos}")

#Numar norocos pentru test : 2415
```


## 18. Verifică dacă un număr este perfect (număr egal cu suma divizorilor săi proprii)

Exemplu:

n = 28 → Output: True (1+2+4+7+14=28)

```python
n = int(input("Introdu un numar natural: "))

suma_divizori = 0

#Verificam toti posibilii divizori pana la jumatatea numarului

for i in range(1, n // 2 +1):
    if n % i == 0 :
        suma_divizori +=i

if suma_divizori == n:
    print(f"{n} este un numar perfect!")
else:
    print(f"{n} nu este un numar perfect!")
```


## 19. Numără literele mari și mici dintr-un text

Exemplu:

"Ana Are Mere" → Output: Upper: 3, Lower: 8

Aici cred ca ai scris gresit la Lower, nu ar trebuii sa fie 7 ? Chiar si daca ai numara spatiul ca si litera mica , atunci ar fi 9 . Sau ???

```python
text = input("Introdu un text: ")

mari = 0
mici = 0

for caracter in text:
    if caracter.isupper()  :  #isupper() verifica daca este litera mare
        mari +=1
    elif caracter.islower():  #islower() verifica daca este litera mica
        mici +=1
print(f"Upper: {mari}, Lower: {mici}")
```


## 20. Scrie un program care inversează cuvintele dintr-o propoziție

Exemplu:

text = "Ana are mere" → Output: "mere are Ana"




```python
text = input("Introdu un text: ")

lista_cuvinte = text.split()   #impartim propozitia in cuvinte

lista_cuvinte.reverse()   #inversam lista folosind metoda reverse

text_inversat = " ".join(lista_cuvinte)   # reconstruim propozitia , join() este opusul lui split(), ea ia o lista si o lipeste intr-un singur text folosind un spearator (in cazul nostru " ")


print("Propozitia introdusa de noi este : " , text)

print("Propozitia inversata este : " , text_inversat)
```
