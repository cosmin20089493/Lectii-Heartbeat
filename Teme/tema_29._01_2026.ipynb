## Contorul uitat din laborator



Într-un laborator vechi de informatică există un dispozitiv misterios: un contor digital. Nimeni nu mai știe exact la ce folosea, dar se știe sigur un lucru: știe doar să numere.

Când apeși un buton, crește. Alt buton, scade. Un buton special adaugă o valoare dorită. Și, dacă te încurci, îl poți reseta la 0.

```python



#Incepem prin a construii o clasa contor

class Contor:
    def __init__(self):       #self.valoare este variabila care tine valoarea contorului
        self.valoare = 0   #fiecare obiect Contor va avea propria valuare

    def inc(self):
        self.valoare += 1  #crestem valoarea contorului cu 1
    def dec(self):
        self.valoare -= 1  #scadem valoarea contorului cu 1
    def add(self, x):
        self.valoare += x   #adauga la contor valoarea x primita ca parametru, unde x este o variabila primita din exterior

    def reset(self):
        self.valoare = 0    #reseteaza contorul de la 0

    def __str__(self):
        return f"Valoarea contorului este : {self.valoare}"

c = Contor()
c.inc()         #  1 +
c.inc()         #  1+ 1 =2
c.add(5)        #  2+5 = 7
c.dec()         # 7-1 = 6
c.add(3)        # 6+3 = 9

print(c)   # aici am facut eu asa sa inteleg logica matematica din spate

c.reset()
c.add(10)

print(c)
```
