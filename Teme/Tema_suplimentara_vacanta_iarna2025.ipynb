# Tema suplimentara vacanta de iarna



Scrie un program care calculează al n-lea element din șirul lui Fibonacci. Șirul este definit astfel:

• F(0) = 0

• F(1) = 1

• F(n) = F(n-1) + F(n-2), pentru n ≥ 2



Rezolvă problema în 3 pași:


## 1. Metoda rudimentară



Folosește doar variabile, condiții și o buclă pentru a obține rezultatul. Programul trebuie să pornească de la F(0) = 0 și F(1) = 1 și să continue până la F(n), unde n este o valoare dată în cod.


"Explicate pentru mine : Șirul Fibonacci funcționează prin acumulare. Dacă pornești cu (0, 0), suma lor va fi mereu 0. Dacă pornești cu (0, 1), pașii vor fi:

$0 + 1 = 1$

$1 + 1 = 2$

$1 + 2 = 3$

$2 + 3 = 5$ ... și tot așa până la 55 (pentru $n=10$). "

```python


n = int(input("Introdu un numar natural: "))

a = 0  #F(0)
b = 1  #F(1)

if n == 0:
    rezultat = a
elif n == 1:
    rezultat = b
else:
    for i in range(2, n+1):
        #calculam urmatorul termen
        urmatorul = a + b
        #actualizam urmatorul termen
        a = b
        b = urmatorul
    rezultat = b

print(f"F({n}) prin metoda rudimentara este: {rezultat}")

# aici am avut rabdare sa incerc pana la 999999 , PC-ul avand nevoie de 14 secunde pentru a calcula totul.
```


## 3. Cu bibliotecă Python



Folosește o funcție gata făcută dintr-o bibliotecă Python pentru a calcula Fibonacci.

• functools.lru_cache aplicat peste funcția recursivă

• sympy.fibonacci(n) din biblioteca SymPy

• orice altă soluție Python care returnează rapid F(n)

```python

from functools import lru_cache

#decoratorul lru_cache transforma o functie lenta intr-una instantanee

@lru_cache(maxsize=None)
def fib_rapid(n):
    if n < 2:
        return n
    return fib_rapid(n-1) + fib_rapid(n-2)

n = int(input("Introdu un numar natural: "))

print(f"F({n}) folosind lru_cache este : {fib_rapid(n)}")

#Aici am observat ca merge pana la n = 1485, daca pun 1486 deja imi da eroare.
```
