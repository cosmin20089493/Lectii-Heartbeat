```python
a = int(input("Introdu un numar a=: "))
b = int(input("Introdu un numar b=: "))




def calculator(a, b):   #intoarce un tuplu (suma, inmultirea, impartirea, rest)

    suma = a + b                              #0
    inmultire = a * b                         #1
    impartire = a / b if b != 0 else None     #2
    rest = a % b if b != 0 else None          #3
    return suma, inmultire, impartire, rest   #4


"""
def inmultire(a, b):
    suma = a * b
    return suma
def inmpartire(a, b):
    suma = a / b
    return suma
def rest(a, b):
    suma = a % b
    return suma
"""


print("Produsul este = " ,calculator(a, b)[1]) #### afisam doar produsul prin index [1]


"""
Mai jos avem metoda "despachetare"
Returnează 4 valori simultan — ele vin împachetate într-un tuplu, adică: ( s, i, d, r ) unde:
s = suma,
i = înmulțirea,
d = împărțirea,
r = restul.
🔹 Ce înseamnă „despachetare”
Când scrii:
_, prod, _, _ = calculator(a, b)
tu „desfaci” acel tuplu în 4 variabile distincte.
Prima valoare (suma) merge în _
A doua valoare (inmultire) merge în prod
A treia și a patra merg iar în _
🔹 Ce face caracterul _
În Python, _ este o convenție — înseamnă „nu am nevoie de această valoare”.
Nu are nimic special, doar ajută la claritate.
Tu ai 4 valori, dar vrei doar a doua (produsul).
Așa că:

_ ignoră rezultatele de care nu ai nevoie,
prod păstrează doar ce vrei tu (rezultatul înmulțirii).
"""


_, prod, _, _ = calculator(a, b)
print(prod)


_,_,_,rest = calculator(a, b)
print(rest)

suma,_,_,_ = calculator(a, b)
print(suma)
```
