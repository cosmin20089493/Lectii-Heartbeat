## Structuri de Date și Algoritmi Esențiali în Python

### 🔹 List – pentru date ordonate

```python
comenzi = ["pizza", "burger", "sushi"]
comenzi.append("paste")
print(comenzi)
print(comenzi[0])

comenzi.sort()  #.sort face sortarea in ordina alfabetica la string-uri. Mai exact se sorteaza Lexicografic ( A>a, B>a....)
print(comenzi)

comenzi.sort(reverse=True)
print(comenzi)
```

### 🔹 Set – pentru unicitate

```python
orase = {"București", "Cluj", "Timișoara"}
orase.add("Iași")
print(orase)
print("Cluj" in orase)  # True
```

### 🔹 Dict – pentru date asociative (key-value)

```python
utilizator = {"nume": "Andrei", "varsta": 28}
utilizator["oras"] = "Cluj"
print(utilizator ["oras"])
```

### 🔹 Tuple – pentru date fixe

```python
coordonate = (45.76, 21.23)

coordonate = coordonate + (4,)

print(coordonate[0])  # 45.76
```
